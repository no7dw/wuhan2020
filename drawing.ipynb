```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

df=pd.read_csv("numbers.csv")
def mm(time_str):
    return datetime.strptime(str(time_str), '%Y%m%d%H')
df['ti']=df['time'].map(mm)

plt.plot( 'ti', 'confirm', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="confirm")
plt.plot( 'ti', 'suspect', data=df, marker='', color='olive', linewidth=2, label="suspect")
plt.plot(  'ti', 'dead', data=df, marker='', color='olive', linewidth=2, linestyle='dashed', label="dead")
plt.legend()
```

```python
df['dead_rate']=df['dead']/df['confirm']
df['heal_rate']=df['heal']/df['confirm']
plt.plot( 'ti', 'dead_rate', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4, label="dead_reate")
plt.plot( 'ti', 'heal_rate', data=df, marker='', color='olive', linewidth=2, label="heal_rate")
plt.legend()
```
